```python
# atividade: criação de um gráfico de linhas que representa os dados mensais de vendas de uma loja fictícia.
# importação da biblioteca
import matplotlib.pyplot as plt

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abirl', 'Maio', 'Junho'] # importação dos valores dos meses de janeiro a junho em array
vendas =[15000, 22000, 18000, 24000, 28000, 21000] # importação dos valores das vendas

# criação do gráfico de linhas
plt.plot(meses, vendas, marker='o', linestyle='-', color='b')

# personalização do gráfico
plt.title('Vendas Mensais') # Título do Gráfico
plt.xlabel('Meses') # xlabel = Meses
plt.ylabel('Valor de Vendas') # ylabel = Valor de Vendas

# função para permitir que as grades (grids) fiquem visíveis
plt.grid(True)

# função para a exibição do gráfico
plt.show()
```
